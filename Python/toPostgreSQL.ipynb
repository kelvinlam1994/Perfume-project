```python
# FOR hktvmall

import pandas as pd
import psycopg2

# Read the CSV file into a DataFrame
df = pd.read_csv('C:/Users/kelvin/Desktop/JDE10/classes/project/13_06_20247HKTVMALL_DATE_TEXT.csv')
# Rename the columns to match the database table
df = df.rename(columns={
    'Title': 'product_name',
    'Price': 'selling_price',
    'Rate': 'rating',
    'brand_name': 'brand',
    'front_note': 'top_note',
    'middle_note': 'heart_note',
    'base_note': 'base_note',
    'Origin': 'Origin',
    'upload_date': 'upload_date',
    'season': 'season',
    'over_one_year_test': 'over_one_year_test'
})

# Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="project",
    user="postgres",
    password="aA27036953"
)

# Create a cursor
cursor = conn.cursor()

# Create the table
create_table_query = """
DROP TABLE IF EXISTS hktvmall;
CREATE TABLE hktvmall (
    product_id SERIAL PRIMARY KEY,
    product_name VARCHAR(255),
    selling_price TEXT,
    rating float,
    brand TEXT,
    top_note TEXT,
    heart_note TEXT,
    base_note TEXT,
    Origin TEXT,
    upload_date text,
    season TEXT,
    over_one_year_test TEXT
);
"""
cursor.execute(create_table_query)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Insert row into PostgreSQL table
    cursor.execute(
        "INSERT INTO hktvmall (product_name, selling_price, rating, brand, top_note, heart_note, base_note, Origin, upload_date, season, over_one_year_test) VALUES (%s, %s, %s, %s, %s, %s, %s, %s, %s, %s, %s)",
        (row['product_name'], row['selling_price'], row['rating'], row['brand'], row['top_note'], row['heart_note'], row['base_note'], row['Origin'], row['upload_date'], row['season'], row['over_one_year_test'])
    )

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
# For SASA   

import pandas as pd
import psycopg2

# Read the CSV file into a DataFrame
df = pd.read_csv('C:/Users/kelvin/Desktop/JDE10/classes/project/sasa_flavor07_06_2024.csv', index_col=[0])

# Rename the columns to match the database table
df = df.rename(columns={
    'Special_Price': 'Selling_Price',
    'top_notes' : 'top_note',
    'heart_notes': 'heart_note',
    'base_notes':'base_note'
    })

# Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="project",
    user="postgres",
    password="aA27036953"
)

# Create a cursor
cursor = conn.cursor()

# Create the table
create_table_query = """
DROP TABLE IF EXISTS SaSa;
CREATE TABLE IF NOT EXISTS Sasa (
    product_id SERIAL PRIMARY KEY,
    product_name TEXT NOT NULL,
    Original_Price TEXT,
    Selling_Price TEXT,
    capacity TEXT,
    brand TEXT,
    top_note TEXT,
    heart_note TEXT,
    base_note TEXT
);
"""
cursor.execute(create_table_query)

# Iterate over each row in the DataFrame
for _, row in df.iterrows():
    # Insert row into PostgreSQL table
    cursor.execute(
        "INSERT INTO sasa (product_name,Original_Price , Selling_Price,capacity,brand,top_note,heart_note,base_note) VALUES (%s, %s, %s, %s, %s, %s, %s, %s)",
        (row[0], row[1], row[2],row[3],row[4], row[5], row[6], row[7])
    )

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```

```python
# Fragrance

import pandas as pd
import psycopg2

# Read the CSV file into a DataFrame
df = pd.read_csv('C:/Users/kelvin/Desktop/JDE10/classes/project/0613fragrance_all.csv')
# Rename the columns to match the database table
df = df.rename(columns={
    'Special_offer label': 'special_offer_label',
    'Title': 'product_name',
    'Top_notes': 'top_note',
    'Heart_notes': 'heart_note',
    'Base_notes' : 'base_note',
    'Price' : 'selling_price'
})

# Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="project",
    user="postgres",
    password="aA27036953"
)

# Create a cursor
cursor = conn.cursor()

# Create the table
create_table_query = """
DROP TABLE IF EXISTS Fragrance;
CREATE TABLE Fragrance (
    product_id SERIAL PRIMARY KEY,
    special_offer_label TEXT,
    product_name TEXT,
    selling_price TEXT,
    top_note TEXT,
    heart_note TEXT,
    base_note TEXT,
    brand TEXT
);
"""
cursor.execute(create_table_query)

# Iterate over each row in the DataFrame
for index, row in df.iterrows():
    # Insert row into PostgreSQL table
    cursor.execute(
        "INSERT INTO Fragrance (special_offer_label, product_name, selling_price, top_note, heart_note, base_note, brand) VALUES (%s, %s, %s, %s, %s, %s, %s)",
        (row[0], row[1], row[2], row[3], row[4], row[5], row[6])
    )

# Commit the changes
conn.commit()

# Close the cursor and connection
cursor.close()
conn.close()
```
